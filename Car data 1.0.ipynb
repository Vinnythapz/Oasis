```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

df = pd.read_csv("https://raw.githubusercontent.com/Vinnythapz/Oasis/main/car%20data.csv")
df
```

```python
from matplotlib import pyplot as plt
df['Present_Price'].plot(kind='line', figsize=(8, 4), title='Present_Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Selling_type'].value_counts()
    for x_label, grp in df.groupby('Fuel_Type')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Fuel_Type')
_ = plt.ylabel('Selling_type')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['Present_Price']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Fuel_Type')):
  _plot_series(series, series_name, i)
  fig.legend(title='Fuel_Type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('Present_Price')
```
